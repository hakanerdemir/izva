```python
#kütüphane

import pandas as pd
```

```python
#veri yükleme

df = pd.read_csv('MacdonellDF.csv')

print(df.to_string()) 
```

```python
#veriyi görüntüleme

print(df)
```

```python
#mean

data = df[['height','finger']].mean()
print(data)
```

```python
#median

data = df[['height','finger']].median()
print(data)
```

```python
#mode

df.mode(axis='columns', numeric_only=True) # bunu sor
```

```python
#yüzde değer

data = df[['height','finger']].quantile([.32,.58,.98])
print(data)
```

```python
#çeyrek değer

data = df[['height','finger']].quantile([0,.25,.50,.75,1])
print(data)
```

```python
#range ve aralik

#max

data = df[['height','finger']].max()
print(data)

data = df[['height','finger']].min()
print(data)

data = df[['height','finger']].max()-df[['height','finger']].min()
print(data)
```
